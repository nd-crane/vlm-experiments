### Install vlm_experiments in Development mode

This repository contains a pyproject.toml generated by [uv](https://docs.astral.sh/uv/). For backwards compatibility there is a requirements.txt that can be used to construct a virtual environment. It is recommended to use a .venv in the local repository directory


```sh
# make sure vlm_experiments package is installed in development mode
$ uv sync

# make changes under nbs/ directory
# ...

# compile to have changes apply to vlm_experiments
$ nbdev_prepare
```
